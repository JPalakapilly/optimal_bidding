```python
import pickle 
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
with open('test_1', 'rb') as pickle_file:
        data = pickle.load(pickle_file)
```

```python
sns.set(style="whitegrid", font_scale=1.8)
x = np.linspace(1000, 1000*len(data[0]), num=len(data[0]))
plt.figure(figsize=(18, 7))
sns.lineplot(x, data[0], label='mean reward')
sns.lineplot(x, data[1], label='max mean reward')
plt.xlabel('episodes')
plt.legend()
plt.savefig('reward_dqn_mpn.png')
```

```python
sns.set(style="whitegrid", font_scale=1.8)
plt.figure(figsize=(10, 5))
x = np.linspace(1000, 1000*len(data[0]), num=len(data[3]))
plt.xlabel('episodes')
plt.ylabel('loss')
sns.lineplot(x, data[3], err_style='band')
plt.savefig('loss_dqn_mpc.png')
```
