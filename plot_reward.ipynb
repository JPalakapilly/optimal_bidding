```python
import pickle 
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
with open('d3331e56-40bb-4328-a7b3-57f0f1a6ec3e.pkl', 'rb') as pickle_file:
        data = pickle.load(pickle_file)
```

```python
x = np.linspace(1000, 1000*len(data[0]), num=len(data[0]))
plt.figure(figsize=(18, 7))
plt.plot(x, data[0], label='mean reward')
plt.plot(x, data[1], label='max mean reward')
plt.grid()
plt.xlabel('timesteps')
plt.legend()
plt.savefig('fig.png')
```
